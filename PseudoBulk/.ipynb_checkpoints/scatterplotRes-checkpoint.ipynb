```python
def plot_scatter_percelltype(method_res,real_res):
    fig,axs = plt.subplots(2,5,figsize=(30,10))
    list_ax = []
    for i in real_res.columns:
        x = []
        y = []
        for j in real_res.index:
            x.append(real_res.loc[j,i])
            y.append(method_res.loc[j,i])
        list_ax.append([x,y,i])
    print(len(list_ax))
    pic = 0
    for ax in axs.flat:
        x = list_ax[pic][0]
        y = list_ax[pic][1]
        ax.scatter(x,y)
        ax.set_xlabel('Real proportion')
        ax.set_ylabel('Estimate')
        ax.set_title(list_ax[pic][2])
        pic += 1
```

```python
def plot_scatter_percelltype_merged(methods_res_list,real_res):
    fig,axs = plt.subplots(2,5,figsize=(30,10))
    list_methods = ['MuSiC','BayesPrism','BLADE','CIBERSORTx','NNLS'] # be careful of the order
    order_method = 0
    for method_res in methods_res_list:
        list_ax = []
        for i in real_res.columns:
            x = []
            y = []
            for j in real_res.index:
                x.append(real_res.loc[j,i])
                y.append(method_res.loc[j,i])
            list_ax.append([x,y,i])
        pic = 0
        for ax in axs.flat:
            x = list_ax[pic][0]
            y = list_ax[pic][1]
            ax.scatter(x,y,label = list_methods[order_method],s=40,alpha=0.7)
            ax.set_xlabel('Real proportion')
            ax.set_ylabel('Estimate')
            ax.legend()
            ax.set_title(list_ax[pic][2])
            lims = [
                np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
                np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes
            ]
            # now plot both limits against each other
            ax.plot(lims, lims, 'k-', alpha=0.7)
#             ax.set_aspect('equal')
            pic += 1
        order_method += 1
```

```python
plot_scatter_percelltype_merged([MuSiC_res,BP_res,BLADE_res,CIBERSORT_res,NNLS_res],real_res)
```
