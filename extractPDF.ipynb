```python
import pandas as pd
import numpy as np
import os,sys
import PyPDF2
import re
```

```python
list_sample = []
list_cellularity = []
dict_new = {}
```

```python
# aim to extract certain text (cellularity) from pdf plots of Kari's project
probe = os.walk("/home/cke/KariPDF")
for path,dir_list,file_list in probe:
    for file_name in file_list:
        name = os.path.join(path,file_name)
        if name.endswith('NEW.pdf'):
            print('find cellularity in',file_name.split("_")[0])
            pdfFileObj = open(name, 'rb')
            pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
            pageObj = pdfReader.getPage(0)
            text = pageObj.extractText()
            cellu = re.findall(r"cellularity \d\.*\d*",text)
            print(cellu)
            dict_new[file_name.split("_")[0]] = cellu[0].split(" ")[1]
```

```python
df_cellularity_new = pd.DataFrame.from_dict(dict_new,orient='index')
df_cellularity_new
```
